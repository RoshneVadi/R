```r
#bargraph using qplot
qplot(mtcars$cyl,geom="bar",fill=I("red"),color=I("black"))
```

```r
qplot(mtcars$cyl,geom="bar",fill=I("black"),color=I("pink"),xlab="Cylinders",ylab="No. of vehicles",main="Cylinders in mtcars")
```

```r
#histograms using qplots
qplot(mtcars$hp,geom="histogram")
```

```r
#binwidth is the width of the bar, 25 is in pixels
#fill using alpha
qplot(mtcars$hp,geom="histogram",binwidth=25,color=I("black"),xlim=c(50,350),xlab="HP",ylab="No. of vehicles",main="HP in mtcars",alpha=I(0))
```

```r
barp<-ggplot(mtcars, aes(x=1, y=sort(carb), fill=sort(carb))) + geom_bar(stat="identity")
print(barp)
```

```r
barp<-barp+coord_polar(theta='y')
print(barp)
```

```r
#theme helper function so that we don't warp the labels and axes with our transformation
barp <- barp + coord_polar(theta='y')

print(barp)
```

```r
barp <- barp + coord_polar(theta='y')
barp <- barp + theme(
    axis.line=element_blank(),
    axis.text.x=element_blank(),
    axis.text.y=element_blank(),
    axis.ticks=element_blank(),
    axis.title.y=element_blank(),
    panel.background=element_blank()) +
    labs(y="Carburetors")
print(barp)
```
